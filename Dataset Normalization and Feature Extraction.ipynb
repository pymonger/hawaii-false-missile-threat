## Data extraction, normalization, and feature extraction on tweets geolocated in Hawaii

### Filter DiscoverText CSV exports of geolocated tweets on 2017-01-14 (control - 1 year before missile threat) and 2018-01-13 (missile threat)

```python
import os, sys, json
import numpy as np
import pandas as pd
from glob import glob
from datetime import datetime
import re


# normalize field names
DT_HEADERS = [
    "Title","Text","UnitId","[M] _twitter_id","[M] actor_image: ",
    "[M] actor_languages: ","[M] actor_links_href: ",
    "[M] actor_listedcount: ","[M] actor_utcoffset: ",
    "[M] favorites_count: ","[M] favoritescount: ",
    "[M] followers_count: ","[M] friends_count: ",
    "[M] geo_coordinates: ","[M] geo_type: ",
    "[M] gnip_urls_expanded_url: ",
    "[M] gnip_urls_expanded_url_description: ",
    "[M] gnip_urls_expanded_url_title: ","[M] hashtag: ",
    "[M] id: ","[M] influence_score: ","[M] inreplyto_link: ",
    "[M] language: ","[M] location_coords: ",
    "[M] location_displayname: ","[M] location_geo_type: ",
    "[M] location_name: ","[M] location_type: ","[M] media_url: ",
    "[M] object_country_code: ","[M] object_posted_time: ",
    "[M] object_twitter_link: ","[M] posted_time: ",
    "[M] real_name: ","[M] retweetcount: ","[M] source: ",
    "[M] statuses_count: ","[M] tweet_id: ","[M] tweet_url: ",
    "[M] twitter_entities_media_description: ",
    "[M] twitter_entities_media_expanded_url: ",
    "[M] twitter_entities_media_source_status_id: ",
    "[M] twitter_entities_media_source_user_id: ",
    "[M] twitter_entities_media_source_user_id_str: ",
    "[M] twitter_entities_media_type: ",
    "[M] twitter_entities_media_url: ",
    "[M] twitter_entities_polls_duration_minutes: ",
    "[M] twitter_entities_polls_end_datetime: ",
    "[M] twitter_entities_polls_options_position: ",
    "[M] twitter_entities_polls_options_text: ",
    "[M] twitter_entities_urls_display_url: ",
    "[M] twitter_entities_urls_url: ",
    "[M] user_bio_summary: ","[M] user_id: ",
    "[M] user_is_verified: ","[M] user_location: ",
    "[M] user_mention: ","[M] user_mention_username: ",
    "[M] user_twitter_page: ","[M] username: ","[M] verb: "
]

DT_HEADERS_FILTER = {
    "Text": "text",
    "[M] _twitter_id": "id",
    "[M] actor_listedcount: ": "user_listed_count",
    "[M] favorites_count: ": "user_favourites_count",
    "[M] followers_count: ": "user_followers_count",
    "[M] friends_count: ": "user_friends_count",
    "[M] geo_coordinates: ": "geo_coords",
    "[M] location_coords: ": "place_coords",
    "[M] location_name: ": "place_name",
    "[M] posted_time: ": "created_at",
    "[M] real_name: ": "user_name",
    "[M] source: ": "source",
    "[M] statuses_count: ": "user_statuses_count",
    "[M] user_id: ": "user_id",
    "[M] user_location: ": "user_location",
    "[M] user_bio_summary: ": "user_description",
    "[M] username: ": "user_screen_name",
    "[M] retweetcount: ": "retweet_count",
    "[M] hashtag: ": "hashtag",
}

def filter_clean(df):
    """Filter columns and cleanup values."""
    
    # filter columns
    df_filtered = df.filter(items=DT_HEADERS_FILTER)
    
    # coerce datetime field to datetime object
    df_filtered['[M] posted_time: '] = pd.to_datetime(df_filtered['[M] posted_time: '])
    
    # extract twitter user id
    df_filtered['[M] user_id: '] = df_filtered['[M] user_id: '].str.replace('id:twitter.com:', '')
    
    # rename to friendlier columns
    df_filtered.rename(columns=DT_HEADERS_FILTER, inplace=True)
    
    # set tweet id as index and string
    df_filtered.set_index('id', inplace=True)
    df_filtered.index = df_filtered.index.map(str)
    
    return df_filtered

def clean_tweets(tweets):
    """Create list of filtered and clean dicts with matching column names."""
    
    all_tweets = []
    filtered_count = 0
    for tweet in tweets:
        # filter quote and retweets
        if tweet['is_quote_status'] or tweet['text'].startswith('RT '):
            filtered_count += 1
            continue
        all_tweets.append({
            "text": tweet['text'].replace('\r\n', ' ').replace('\n', ' '),
            "id": str(tweet['id']),
            "user_listed_count": tweet.get('user', {}).get('listed_count', 0),
            "user_favourites_count": tweet.get('user', {}).get('favourites_count', 0),
            "user_followers_count": tweet.get('user', {}).get('followers_count', 0),
            "user_friends_count": tweet.get('user', {}).get('user_friends_count', 0),
            "geo_coords": np.nan if tweet['coordinates'] is None else "%f, %f" %
                (tweet['coordinates']['coordinates'][1],
                 tweet['coordinates']['coordinates'][0]),
            "place_coords": np.nan if tweet['place'] is None else tweet['place']['bounding_box']['coordinates'][0],
            "place_name": np.nan if tweet['place'] is None else tweet['place']['name'],
            "created_at": pd.to_datetime(tweet['created_at']),
            "user_name": tweet['user']['name'],
            "source": re.search(r'>(.+)<', tweet['source']).group(1),
            "user_statuses_count": tweet.get('user', {}).get('statuses_count', 0),
            "user_id": tweet['user']['id'],
            "user_location": tweet['user']['location'],
            "user_description": tweet['user']['description'].replace('\r\n', ' ').replace('\n', ' ') if tweet['user']['description'] is not None else None,
            "user_screen_name": tweet['user']['screen_name'],
            "retweet_count": tweet['retweet_count'],
            "hashtag": "; ".join([i['text'] for i in tweet['entities']['hashtags']]),
        })
    print("total: %d" % len(all_tweets))
    print("filtered: %d" % filtered_count)
    df = pd.DataFrame(all_tweets)
    df.set_index('id', inplace=True)
    df.index = df.index.map(str)
    return df
```

```python
# read csv into data frame
df_20180113 = pd.read_csv('data/discovertext/hawaii_missile_crisis-export-20180221-113313.csv',
                          infer_datetime_format=True)
df_20180113.head()
```

```python
# add data from full archive premium search on place=", HI"
with open('data/twitter/premium_search-full/hi_tweets-full-place.json') as f:
    df_20180113_prem = clean_tweets(json.load(f))

# add unseen tweets
print("HI 2018-01-13 tweets from discovertext: {}".format(len(df_20180113_filtered)))
print("HI 2018-01-13 tweets from premium: {}".format(len(df_20180113_prem)))
new = 0
old = 0
for id, row in df_20180113_prem.iterrows():
    if id not in df_20180113_filtered.index: new += 1
    else: old += 1
    df_20180113_filtered.loc[id] = row
print("premium tweets added: {}".format(new))
print("premium tweets not added: {}".format(old))
df_20180113_filtered = df_20180113_filtered.sort_values(by=['created_at'])
print("total unique HI 2018-01-13 tweets from discovertext and premium: {}".format(len(df_20180113_filtered)))
df_20180113_filtered.head()
```

### Filter and clean JSON results of geolocated tweets from the 30-day premium twitter API for 2018-03-03T00:00:00 through  2018-03-03T23:59:59 (control - 1.5 months after missile threat)

```python
from glob import glob
import re


# twitter api limits geolocation search of tweets to a point with max radius of 25 miles
# therefore I had to make multiple geolocation searches (1 per island centered on the 
# approximate center of the island)
tweet_files = glob('data/twitter/hawaii_missile_crisis-control2-fullday/*.json')
all_tweets = []
for tweet_file in tweet_files:
    print("processing {}".format(tweet_file))
    with open(tweet_file) as f:
        all_tweets.extend(json.load(f))
        
df_20180303_filtered = clean_tweets(all_tweets)
df_20180303_filtered.to_csv('data/hawaii_missile_crisis_control-20180303.csv')
print("HI 2018-03-03 tweets from premium: {}".format(len(df_20180303_filtered)))
df_20180303_filtered.head()
```

```python
# merge data frames into one and dump to CSV
frames = [ df_20180303_filtered, df_20170114_filtered, df_20180113_filtered ]
all_df = pd.concat(frames).sort_values(by=['created_at'])
all_df.to_csv('data/hawaii_missile_crisis-all.csv')
print(len(all_df))
all_df = all_df[~all_df.index.duplicated(keep='first')]
print(len(all_df))
all_df.head()
```

### Add class label: mortality_salience

This stage is a manual task. Here we add a new column named mortality_salience. It will either take a value of 1 or 0. If 1, we label that the tweeter is experiencing mortality salience. Here we defer to the domain expertise on mortality salience which defines that a person experiences mortality salience when they are coping with imminent death. In this task, I:

1. label all tweets within the crisis period (geolocated in HI between 2018-01-13 08:07-08:21AM HST) as "1" and all others as "0"
1. Manually go through the tweets during this crisis period and change any tweets not related to the missile crisis to "0" (e.g. twitpics about food, vacationing in paradise, etc.)
1. Go through tweets after the crisis period ends (after 08:20AM which is the first official notification of the cancellation of the missile threat) and label any tweets as "1" where I determine that the tweeter may not be aware of the cancellation yet

```python
# add column for mortality_salience (default to 0)
all_df_labelled = all_df.assign(mortality_salience=0)

# set tweets that fall within crisis period to 1
all_df_labelled.loc[(all_df_labelled['created_at'] >= datetime(2018, 1, 13, 18, 7, 0)) &
                    (all_df_labelled['created_at'] <= datetime(2018, 1, 13, 18, 21, 0)),
                    'mortality_salience'] = 1

# save
all_df_labelled.to_csv('data/hawaii_missile_crisis-all-labelled.csv')
all_df_labelled.head()

# open "hawaii_missile_crisis-all-labelled.csv" in excel and 
# update labels based on #2 and #3 above. Save.
```
